## Exploratory Analysis and Visualization

### Most Wins by Teams

```python
mw=most_wins.iloc[ : ,[0,-1]]
mw.rename(columns={'Match # 144':'Wins'}, inplace = True)
mw=mw.sort_values(by=['Wins'], ascending=False).reset_index()
mw.iloc[:,[1,-1]]   #else it will also includes prev indexes
```

```python
#plt.figaspect
#plt.figure(figsize=(20,10))
plt.figure()
plt.grid()
plt.title('Most Wins by Teams')
sns.barplot(x='Wins',y='Team Name',data=mw)
```

```python
mwt=most_wins.transpose()   #for line plot
nh=mwt.iloc[0]
mwt=mwt[1:]
mwt.columns=nh
```

```python
#plot=mwt.plot(figsize=(20, 10))
plot=mwt.plot()
plot.axes.get_xaxis().set_visible(False)
plot.grid()
plot.legend(loc=0, prop={'size': 8})
plot
```

### Most Sixes by Players

```python
ms=most_sixes.iloc[ : ,[0,-1]]
ms.rename(columns={'PSL 2020 | Match # 34':'Sixes'}, inplace = True)
ms=ms.sort_values(by=['Sixes'], ascending=False).reset_index()

del ms['index']
ms
#ms.iloc[:,[1,-1]] 
```

#### Top 15 Players with most Sixes

```python
tms=ms.head(15)
tms
```

```python
plt.figure()
plt.grid()
plt.title('Top 15 Players with most sixes')
sns.barplot(x='Sixes',y='Player Name',data=tms)
```

### Most Wickets taken by a Player

```python
mw=most_wickets.iloc[ : ,[0,-1]]
mw.rename(columns={'PSL 2020 | Match # 34':'Wickets'}, inplace = True)
mw=mw.sort_values(by=['Wickets'], ascending=False).reset_index()

del mw['index']
mw
```

#### Top 15 Players with most Wickets

```python
tmw=mw.head(15)
tmw
```

```python
plt.figure()
plt.grid()
plt.title('Top 15 Players with most wickets')
sns.barplot(x='Wickets',y='Player Name',data=tmw)
```

### Player(s) in Top 20 for both Sixes & Wickets

```python
tms['Player Name'].isin(tmw['Player Name']).value_counts()
```

```python
best=tms.loc[tms['Player Name'].isin(tmw['Player Name'])].reset_index()
best=best['Player Name']
best=best.to_string(index=False)

best
```

### Matches Info for runs above 180

```python
#Split Score
highest_total[['Runs','Out']]=highest_total.Score.str.split(pat='/',expand=True).astype('int')
```

```python
highest_total
```

```python
plt.plot(highest_total.Runs)
```

#### Which Innings Players has the highest Out Rate?

```python
ivo=highest_total.iloc[ : ,[4,-1]]
```

```python
ivo.Inns.replace({
    1:'Batting',
    2:'Fielding'
},inplace=True)
ivo=ivo.groupby("Inns")
ivo=ivo.sum()

ivo
```

#### Which innings have the most highest scores? 

```python
inn = highest_total.pivot_table(columns=['Inns'], aggfunc='size')
inn
```

#### Which venue hosted most of the matches? 

```python
highest_total.Ground.unique()
```

```python
venue=highest_total.groupby('Ground')[['Match Date']].count()
venue=venue.sort_values('Match Date',ascending=False).reset_index()
venue.rename(columns={
    'Ground':'Venue',
    'Match Date' : 'Total'
},inplace=True)
venue
```

```python
labels=list(venue.Venue)
```

```python
plt.pie(venue.Total, labels=labels)
```
